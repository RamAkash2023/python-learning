```python
def convertRuns(string):
    if str(string).count('*')>=1:
        return float(str(string).replace('*',''))
    elif str(string)=='DNB' or str(string)=='TDNB':
        return float(0)
    else:
        return float(string)
def convertFours(string):
    if str(string)=='-':
        return float(0)
    else:
        return float(string)
df['Runs']=df['Runs'].apply(convertRuns)
df['4s']=df['4s'].apply(convertFours)
```
